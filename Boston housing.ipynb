```python
#Task 3: Hypothesis Testing & Statistical Analysis
```

```python
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
#T-test: Does the Charles River Affect House Prices?
# Split MEDV by CHAS
chas_1 = boston_df[boston_df['CHAS'] == 1]['MEDV']
chas_0 = boston_df[boston_df['CHAS'] == 0]['MEDV']

# Perform independent samples t-test
t_stat, p_val = stats.ttest_ind(chas_1, chas_0, equal_var=False)
print("T-statistic:", t_stat)
print("P-value:", p_val)
```

```python
#2) ANOVA: Does House Age Affect Median Value?
# Use the same age groups from Task 2
bins = [0, 35, 70, 100]
labels = ['≤35 years', '35–70 years', '>70 years']
boston_df['AGE_GROUP'] = pd.cut(boston_df['AGE'], bins=bins, labels=labels, right=True)

# Perform one-way ANOVA
model = ols('MEDV ~ C(AGE_GROUP)', data=boston_df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
#Correlation: Relationship between NOX and INDUS
corr_coeff, p_val = stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
print("Correlation coefficient:", corr_coeff)
print("P-value:", p_val)
```

```python
#Regression: Effect of Distance to Employment Centres (DIS) on House Prices
X = sm.add_constant(boston_df['DIS'])  # Add constant for intercept
y = boston_df['MEDV']

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
## Conclusions

'''1. **Charles River Proximity:**  
   Homes near the Charles River are statistically more expensive.

2. **Age of Houses:**  
   Newer houses (≤35 years) have higher median values than older ones.

3. **Industrialization & Pollution:**  
   Strong positive correlation between industrial area (`INDUS`) and air pollution (`NOX`).

4. **Distance to Employment Centers:**  
   Greater distance (`DIS`) positively influences house value — indicating suburban areas tend to be pricier.'''

```
